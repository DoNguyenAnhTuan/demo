## Visualize Validation Results

```python
import matplotlib.pyplot as plt
import glob as glob
```

```python
%cd /content/drive/MyDrive/Faster_rcnn/fastercnn-pytorch-training-pipeline/outputs/training/custom_training
results_dir_path = '/content/drive/MyDrive/Faster_rcnn/fastercnn-pytorch-training-pipeline/outputs/training/custom_training'
valid_images = glob.glob(f"{results_dir_path}/*.jpg")

for i in range(3):
    plt.figure(figsize=(10, 7))
    image = plt.imread(valid_images[i])
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```
